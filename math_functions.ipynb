Math Functions

```python
import matplotlib.pyplot as plt
import numpy as num
```

**Step 1: Create x values**

```python
x_list = list(num.arange(-10, 11, 1))
print(x_list)
```

**Step 2: create functions**

```python
y_list = []
for i in x_list:
    y_list.append(i)

y_list2 = [-i for i in x_list]
print(y_list2)
```

Pythonic

**Step 3: plot the graph**

```python
fig, ax = plt.subplots() 
ax.plot(x_list, y_list2)
```

This looks like something created in Excel or Google Sheets, not Python

**Step 4: make the chart look pretty**

```python

def format_chart(fig, ax, title='a chart', x_title='X value', y_title='Y value', size=6):
    ax.set_title(title)
    ax.set_xlabel(x_title)
    ax.set_ylabel(y_title)
    ax.grid(True, which='both', linestyle='--', linewidth=0.4) 
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5) 
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    fig.set_size_inches(size, size)
```

Fignomenal

```python

fig, ax = plt.subplots()
ax.plot(x_list, y_list2)
format_chart(fig, ax, 'This is a chart created in Python, not Excel or Google Sheets')
plt.show()
```

```python
y_list3 = [i+2 for i in x_list]

fig, ax = plt.subplots()
ax.plot(x_list, y_list3)
format_chart(fig, ax, 'This is a chart created in Python, not Excel or Google Sheets')
plt.show()
```

```python
y_list4 = [(0.5*i)+2 for i in x_list]

fig, ax = plt.subplots()
ax.plot(x_list, y_list4)
format_chart(fig, ax, 'This is a chart created in Python, not Excel or Google Sheets')
plt.show()
```

```python
y_list5 = [-10 + (i**2) for i in x_list]

fig, ax = plt.subplots()
ax.plot(x_list, y_list5)
format_chart(fig, ax, 'This is a chart created in Python, not Excel or Google Sheets')
plt.show()
```
